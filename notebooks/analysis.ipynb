# ToxiPredict AI - Exploratory Data Analysis

This notebook explores the Tox21 dataset, covering data loading, preprocessing, EDA, and model selection for drug toxicity prediction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Loading
df = pd.read_csv('../dataset/tox21.csv')
df.head()
```

```python
# Inspect dataset structure
df.info()
print("Missing values:\n", df.isnull().sum())
```

```python
# 2. EDA: Toxicity Distribution
sns.countplot(x='Toxicity', data=df)
plt.title('Distribution of Toxicity')
plt.show()
```

```python
# 3. EDA: Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
# 4. Feature Pairplot
sns.pairplot(df, hue='Toxicity', vars=['MolecularWeight', 'LogP', 'TPSA'])
plt.show()
```
